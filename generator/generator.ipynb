```python
import folium
import json

my_map = folium.Map(location=[53.261382, 50.240032],
                tiles='openstreetmap', # оно такое по умолчанию
                zoom_start=13)

with open('config.json') as config:
    config_data = json.load(config)
    
    folium.Marker(
    location=[config_data['start']['latitude'], config_data['start']['longitude']],
    tooltip='Старт',
    icon=folium.Icon(color="red")).add_to(my_map)
    
    
for point in config_data['points']:
    folium.Marker(
        location=[point['latitude'], point['longitude']],
        tooltip='Промежуточный пункт',
        icon=folium.Icon(color="green")).add_to(my_map)

with open('output.json') as points:
    points = json.load(points)

for point in points:
    folium.Marker(
        location=[point['latitude'], point['longitude']],
        tooltip=[point['latitude'], point['longitude']]).add_to(my_map)
my_map
```
